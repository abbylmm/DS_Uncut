# Specially For Avocado Lovers 🥑🥑🥑

Historical Data of Avocados in U.S. (Source: [Kaggle](https://www.kaggle.com/neuromusic/avocado-prices))

Objectives:
* Avocado Price Distribution
* Price vs Years
* Price vs Type
* Price by Region
* Correlation between Price and Volume


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pylab as pl
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/abbylmm/DS_Uncut/main/avocado.csv")
df.head()
```

```python
df.info()
```

# Avocado Price Distribution

```python
pl.figure(figsize=(10,5))
pl.title("Price Distribution")
avocado = sns.distplot(df["AveragePrice"], color = 'r')
```

* Distribution seems like a bell curve but not perfectly normal
* Most avocados are close to the average price of USD 1.0 - 1.5

# Average Price vs Years

```python
sns.set(rc={'figure.figsize':(5,5)})
g = sns.boxplot(x="year", y="AveragePrice", data=df)
g.set_xticklabels(g.get_xticklabels(), rotation=90)
```

*   Price of avocados in Year 2017 is highest compared with other years
*   Reason: Avocados “are experiencing a 'down' year in 2017” because of the drought in California (Source: [CNN](https://money.cnn.com/2017/09/09/news/economy/avocado-prices-soar/index.html))

# Average Price vs Type

```python
sns.set(rc={'figure.figsize':(5,5)})
g = sns.boxplot(x="type", y="AveragePrice", data=df)
g.set_xticklabels(g.get_xticklabels(), rotation=90)
```

```python
cols = ['type', 'AveragePrice', 'Total Volume', 'Total Bags']
df2 = df.loc[:,cols]
df2.groupby('type').mean()
```

* Organic Avocados are more expensive than Conventional Avocados
* We all like natural products and willing to pay a higher price for it 😝

# Average Price by Region

```python
sns.set(rc={'figure.figsize':(12,12)})
g = sns.boxplot(x="region", y="AveragePrice", data=df)
g.set_xticklabels(g.get_xticklabels(), rotation=90)
```

```python
avg_price_by_region = df.groupby("region")["AveragePrice"].mean().reset_index().sort_values("AveragePrice")
print(avg_price_by_region)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
avg_price_by_region[:10].plot(kind="bar", x="region", y="AveragePrice", title="Top 10 Region with Cheapest Avocado Price")
plt.show()
```

Hurray! Citiest with Cheap Avocado to have Avocado Party 🎉
* Houston
* Dallas-Fort Worth
* South Central

# Correlation between Price and Volume

```python
cols = ['AveragePrice', 'Total Volume', 'Total Bags', '4046', '4225', '4770']
df3 = df.loc[:,cols]
df3.head()
```

```python
corr = df3.corr()
corr
```

```python
sns.set(rc = {'figure.figsize':(5,5)})
sns.heatmap(corr, annot = True)
plt.title('Correlation Matrix', size=16)
plt.show()
```

*   Negative correlated [-0.19] between average price and total volume
*   The lower the availability of avocados, the higher the price
