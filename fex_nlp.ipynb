# TF-IDF

* Term Frequency
  - How frequently a term appears in the entire document
* Inverse Document Frequency
  - Whether a term is rare or frequent across the documents in the entire corpus
* [Reference](https://www.geeksforgeeks.org/feature-extraction-techniques-nlp/)

# Cosine Similarity

* A metric used to measure how similar the documents are irrespective of their size
* [Reference](https://www.machinelearningplus.com/nlp/cosine-similarity/#2whatiscosinesimilarityandwhyisitadvantageous)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
feature = tfidf.fit_transform(df['description'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
feature = tfidf.fit_transform(df['description'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
q1 = "nuclear war"
q2 = "artificial intelligence"
q3 = "new york"
q4 = "detective agency"
q5 = "train to busan"

queries = [q1, q2, q3, q4, q5]

for i in queries: 
    query_feature = tfidf.transform([i])

    if i == q1: 
        print("\n\nKeywords 1:",'\033[1m' + q1 + '\033[0m')
        print(query_feature)
    elif i == q2: 
        print("\n\nKeywords 2:",'\033[1m' + q2 + '\033[0m')
        print(query_feature)
    elif i == q3: 
        print("\n\nKeywords 3:",'\033[1m' + q3 + '\033[0m')
        print(query_feature)
    elif i == q4: 
        print("\n\nKeywords 4:",'\033[1m' + q4 + '\033[0m')
        print(query_feature)
    elif i == q5: 
        print("\n\nKeywords 5:",'\033[1m' + q5 + '\033[0m')
        print(query_feature)

    cosims = cosine_similarity(query_feature, feature).flatten()
    results = reversed(cosims.argsort()[-6:-1])
    # print(results)

    for r in results: 
        print('-----')
        print(f"similarity score: {cosims[r]:.4f}, \ndescription: {df.iloc[r]['description']}")
```

(a, b) c

* a is document id
* b is token id
* c is IDF score
